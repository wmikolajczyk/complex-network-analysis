```python
# IMPORTS
import networkx as nx
import matplotlib.pyplot as plt

from compare_graphs import compare
from generate_graph_nn import attach_attributes, get_trained_model, generate_graph_by_nn
```

```python
# GENERATE GRAPHS (Erdos-Renyi, Watts-Strogatz and Albert-Barabasi), CALCULATE NODE ATTRIBUTES AND ATTACH THESE ATTRIBUTES
seed_num = 93
n1, p1 = 30, 0.6
erdos_renyi_graph = nx.erdos_renyi_graph(n1, p1, seed=seed_num) # it's just gnp_random_graph
attach_attributes(erdos_renyi_graph)

n2, k2, p2 = 30, 2, 0.6
watts_strogatz_graph = nx.watts_strogatz_graph(n2, k2, p2, seed=seed_num)
attach_attributes(watts_strogatz_graph)

n3, m3 = 30, 2
albert_barabasi_graph = nx.barabasi_albert_graph(n3, m3, seed=seed_num)
attach_attributes(albert_barabasi_graph)
```

```python
# CALCULATE AVG NUM EDGES FOR PRIORITY RANK, TRAIN MODEL, GENERATE NEW GRAPH BY NN USING TRAINED MODEL
erdos_avg_num_edges = round(erdos_renyi_graph.number_of_edges()/erdos_renyi_graph.number_of_nodes())
erdos_renyi_model = get_trained_model(erdos_renyi_graph)
new_erdos_renyi_graph = generate_graph_by_nn(erdos_renyi_model, erdos_renyi_graph, erdos_avg_num_edges)

watts_avg_num_edges = round(watts_strogatz_graph.number_of_edges()/watts_strogatz_graph.number_of_nodes())
watts_strogatz_model = get_trained_model(watts_strogatz_graph)
new_watts_strogatz_graph = generate_graph_by_nn(watts_strogatz_model, watts_strogatz_graph, watts_avg_num_edges)

albert_avg_num_edges = round(albert_barabasi_graph.number_of_edges()/albert_barabasi_graph.number_of_nodes())
albert_barabasi_model = get_trained_model(albert_barabasi_graph)
new_albert_barabasi_graph = generate_graph_by_nn(albert_barabasi_model, albert_barabasi_graph, albert_avg_num_edges)
```

```python
# DRAW ORIGINAL GRAPHS
nx.draw(erdos_renyi_graph, with_labels=True)
plt.show()
nx.draw(watts_strogatz_graph, with_labels=True)
plt.show()
nx.draw(albert_barabasi_graph, with_labels=True)
plt.show()
```

```python
# DRAW GENERATED GRAPHS
nx.draw(new_erdos_renyi_graph, with_labels=True)
plt.show()
nx.draw(new_watts_strogatz_graph, with_labels=True)
plt.show()
nx.draw(new_albert_barabasi_graph, with_labels=True)
plt.show()
```

```python
# GET COMPARE DICTS FOR ORIGINAL AND GENERATED GRAPHS
compare_erdos_renyi = compare(erdos_renyi_graph, new_erdos_renyi_graph)
compare_watts_strogatz = compare(watts_strogatz_graph, new_watts_strogatz_graph)
compare_albert_barabasi = compare(albert_barabasi_graph, new_albert_barabasi_graph)
```

```python
# PRINT ORIGINAL AND GENERATED GRAPH COMPARSION RESULTS
print('Erdos-Renyi comparison')
[print('{}: {}'.format(k, v)) for k, v in compare_erdos_renyi.items()]
print('\nWatts-Strogatz comparison')
[print('{}: {}'.format(k, v)) for k, v in compare_watts_strogatz.items()]
print('\nAlbert-Barabasi comparison')
[print('{}: {}'.format(k, v)) for k, v in compare_albert_barabasi.items()]
```
