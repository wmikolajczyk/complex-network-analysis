```python
import os
import shutil

from real_graph import get_available_datasets, download_dataset, extract_dataset, count_attributes
```

```python
available_datasets = get_available_datasets()
```

```python
result_dict = {}
for dataset_name in available_datasets:
    archive_filepath = download_dataset(dataset_name, available_datasets)
    if archive_filepath:
        network_filepath = extract_dataset(dataset_name, archive_filepath)
        num_of_attrs = count_attributes(network_filepath)
        result_dict[dataset_name] = num_of_attrs
        if num_of_attrs > 0:
            print('{} :{}'.format(dataset_name, num_of_attrs))
        else:
            # delete archive_filepath
            os.remove(archive_filepath)
            # delete network_filepath
            shutil.rmtree(network_filepath)
    
```

```python
import operator

results_with_attrs = {k: v for k, v in result_dict.items() if v > 0}
sorted_results = sorted(results_with_attrs.items(), key=operator.itemgetter(1), reverse=True)
sorted_results
```
